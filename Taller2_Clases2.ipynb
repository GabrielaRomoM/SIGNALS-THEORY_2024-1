# EJERCICIO 1: Sistema de Biblioteca

Crea una clase Libro que contenga atributos como título, autor, año de publicación y estado (disponible o prestado). Luego, desarrolla una clase Biblioteca que gestione una colección de objetos Libro. Esta clase debe incluir métodos para:

Añadir un nuevo libro al catálogo.Prestar un libro a un usuario (cambiar el estado a prestado).Devolver un libro (cambiar el estado a disponible). Buscar libros por autor o título. Mostrar los libros disponibles para préstamo.


```python
class libro:
  def __init__(self,titulo,autor,year,estado):
      self.titulo = titulo
      self.autor = autor
      self.year = year
      self.estado = estado

  def __str__(self):
      return f"{self.titulo} - {self.autor}, ({self.year}), [{self.estado}]"

class Biblioteca:
  def __init__(self):
      self.catalogo = []

  def agg(self,libro):
      self.catalogo.append(libro)

  def prestar(self,titulo):
      for libro in self.catalogo:
        if libro.titulo == titulo and libro.estado == "disponible":
          libro.estado= "prestado"
          print(f"El libro {titulo} se ha prestado.")
          return
      print(f"El libro {titulo} no esta disponible.")

  def devolver(self,titulo):
      for libro in self.catalogo:
        if libro.titulo == titulo and libro.estado == "prestado":
          libro.estado= "disponible"
          print(f"El libro {titulo} esta nuevamente disponible.")
          return
      print(f"El libro {titulo} no se puede devolver ya que se encuentra disponible.")


  def buscarautor(self,autor):
      autorb = [libro for libro in self.catalogo if libro.autor == autor]
      if autorb:
            for libro in autorb:
                print(libro)
      else:
            print("No se encontraron libros de ese autor.")

  def buscartitulo(self,titulo):
      titulob = [libro for libro in self.catalogo if libro.titulo == titulo]
      if titulob:
           for libro in titulob:
                print(libro)
      else:
           print("No se encontraron libros con ese titulo.")

  def mostrar(self):
      disponibilidad = [libro for libro in self.catalogo if libro.estado == "disponible"]
      if disponibilidad:
            for libro in disponibilidad:
                print(libro)
      else:
            print("No hay libros disponibles en este momento.")

biblioteca = Biblioteca()

cantlibros = int(input("Ingrese el número de libros que desea registrar: "))

for i in range(cantlibros):
    titulo = input("Ingrese el título del libro: ")
    autor = input("Ingrese el autor del libro: ")
    year = input("Ingrese el año de publicación del libro: ")
    estado = input("Ingrese el estado del libro: ")
    libronew = libro(titulo, autor, year,estado)
    biblioteca.agg(libronew)

while True:
 menu= input("\n(1) Ver menu de opciones (0)Salir: ")
 if menu=='1':
  print("\n OPCIONES:")
  print("\n(1) Añadir un nuevo libro\n(2) Prestar un libro\n(3) Devolver un libro\n(4) Buscar un libro\n(5) Mostrar libros disponibles\n(0) Salir\n")
  L = input("¿Qué desea realizar?: ")

  if L=='1':
    while True:
      ln= input("\n¿Desea añadir un nuevo libro? SI(a) NO(b): ")
      if ln=='a':
       titulo = input("Ingrese el título del libro: ")
       autor = input("Ingrese el autor del libro: ")
       year = input("Ingrese el año de publicación del libro: ")
       estado = input("Ingrese el estado del libro: ")
       libronew = libro(titulo, autor, year,estado)
       biblioteca.agg(libronew)
      else:
       print("\nSe finalizó exitosamente el registro.")
       break
  elif L=='2':
    titulo = input("Ingrese el título del libro que desea prestar: ")
    biblioteca.prestar(titulo)
  elif L=='3':
    titulo = input("Ingrese el título del libro que desea devolver: ")
    biblioteca.devolver(titulo)
  elif L== '4':
     s= input("¿Desea buscar el libro por autor(a) o por titulo(b): ")
     if s=='a':
       autor = input("Ingrese el autor que desea buscar: ")
       biblioteca.buscarautor(autor)
     else:
       titulo = input("Ingrese el título que desea buscar: ")
       biblioteca.buscartitulo(titulo)
  elif L=='5':
   biblioteca.mostrar()
  elif L == '0':
        print("Se ha finalizado el programa")
        break
  else:
   print("Entrada no valida.")
 else:
    print("Se ha finalizado el programa")
    break
```

# EJERCICIO 3: Sistema de Gestión de Empleados

Implementa una clase Empleado con atributos como nombre, identificación, puesto y salario. Posteriormente, desarrolla una clase Departamento que contenga una lista de empleados. Esta clase debe ser capaz de:

Añadir un nuevo empleado. Eliminar un empleado por su identificación.
Aumentar el salario de un empleado por un porcentaje dado. Mostrar la información de todos los empleados. Calcular el gasto total en salarios dentro del departamento.


```python
class empleado:
  def __init__(self,nombre,id,puesto,salario):
      self.nombre = nombre
      self.id = id
      self.puesto = puesto
      self.salario = salario

  def __str__(self):
    return f"{self.nombre}- {self.id}, Puesto: {self.puesto}, Salario: [{self.salario}]"

class Departamento:
  def __init__(self):
      self.empleados = []

  def nuevo(self,empleado):
      self.empleados.append(empleado)

  def eliminar(self,id):
      for empleado in self.empleados:
        if empleado.id == id:
          self.empleados.remove(empleado)
          print(f"El empleado con ID N° {id} ha sido eliminado correctamente.")
          return
      print(f"El empleado con ID N° {id} no existe.")

  def aumento(self,id,percent):
     for empleado in self.empleados:
          if empleado.id == id:
              aumento = (empleado.salario * (percent / 100))
              empleado.salario = empleado.salario + aumento
              print(f"El nuevo salario del empleado con ID N°{id} es de {empleado.salario}.")

  def mostrar(self):
      info = [empleado for empleado in self.empleados]
      if info:
            for empleado in info:
                print(empleado)
      else:
            print("No hay empleados en este momento.")

  def gasto(self):
      gastotal = sum(empleado.salario for empleado in self.empleados)
      print(f"El gasto total en salarios del departamento es: {gastotal}")


depart = Departamento()

cantemps = int(input("Ingrese el número de empleados que desea registrar: "))

for i in range(cantemps):
    nombre = input("Ingrese el nombre del empleado: ")
    id = input("Ingrese el número de identificación (ID) del empleado: ")
    puesto = input("Ingrese el puesto/cargo que tiene el empleado: ")
    salario = float(input("Ingrese el salario del empleado: "))
    empnew = empleado(nombre,id,puesto,salario)
    depart.nuevo(empnew)

while True:
 opp= input("\n(1) Ver menu de opciones (0) Salir: ")
 if opp=='1':
  print("\n OPCIONES:")
  print("\n(1) Añadir un nuevo empleado\n(2) Eliminar un empleado\n(3) Aumentar salario")
  print("(4) Mostrar información de todos los empleados\n(5) Calcular gasto total del departamento(salarios)\n(0) Salir\n")
  E = input("¿Qué desea realizar?: ")

  if E =='1':
    while True:
      em= input("\n¿Desea añadir registrar un nuevo empleado? SI(a) NO(b):")
      if em=='a':
       nombre = input("Ingrese el nombre del empleado: ")
       id = input("Ingrese el número de identificación (ID) del empleado: ")
       puesto = input("Ingrese el puesto/cargo que tiene el empleado: ")
       salario = float(input("Ingrese el salario del empleado: "))
       empnew = empleado(nombre,id,puesto,salario)
       depart.nuevo(empnew)
      else:
        print("\nSe ha finalizado correctamente el registro.")
        break
  elif E =='2':
    id = input("Ingrese el numero de identificacion del empleado que desea eliminar: ")
    depart.eliminar(id)
  elif E =='3':
    id = input("Ingrese el numero de identificacion del empleado que va a tener un aumento en su salario: ")
    percent= float(input("¿En qué porcentaje aumentará el salario?: "))
    depart.aumento(id,percent)
  elif E == '4':
    depart.mostrar()
  elif E =='5':
    depart.gasto()
  elif E == '0':
        print("Se ha finalizado el programa")
        break
  else:
   print("Entrada no valida.")
 else:
   print("Se ha finalizado el programa")
   break
```

# EJERCICIO 4: Tienda en Línea
Crea una clase Producto con atributos tales como identificador, nombre, precio y cantidad en stock. Después, diseña una clase CarritoDeCompras que permita a los usuarios agregar o eliminar productos, y calcular el total de su compra. La clase CarritoDeCompras debe ofrecer métodos para:

Añadir un producto al carrito. Eliminar un producto del carrito. Mostrar los productos en el carrito. Calcular el total de la compra. Aplicar descuentos a productos específicos o al total de la compra.


```python
class Producto:
  def __init__(self,codigo,nombre,precio,stock):
      self.codigo = codigo
      self.nombre = nombre
      self.precio = precio
      self.stock = stock

  def __str__(self):
    return f"{self.codigo}- {self.nombre}, Precio: {self.precio}, Cantidad en stock: [{self.stock}]"

productosdispo =[
      Producto("001","Anillo",20000,10),
      Producto("002","Pulsera",30000,15),
      Producto("003","Collar",25000,20),
      Producto("004","Aretes",30000,15),
      Producto("005","Diadema",15000,12)
      ]

class carritodecompras:
  def __init__(self):
      self.productos = []

  def nuevo(self,codigo,stock):
       for producto in productosdispo:
          if producto.codigo == codigo:
              if producto.stock >= stock:
                  prodnew = Producto(producto.codigo, producto.nombre, producto.precio, stock)
                  self.productos.append(prodnew)
                  print(f"Se ha añadido {stock} unidad(es) del producto {prodnew.nombre} al carrito.")
                  producto.stock -= stock
                  return
              else:
                  print(f"No hay suficiente stock disponible para el producto {producto.nombre}.")
                  return
       else:
        print("El producto no está disponible en la tienda.")

  def eliminar(self,codigo):
      for producto in self.productos:
         if producto.codigo == codigo:
            if producto.stock > 0:
                 producto.stock -= 1
                 print(f"EL producto {codigo}-{nombre} ha sido eliminado del carrito. Stock actualizado: {producto.stock}")
            else:
                 self.productos.remove(producto)
                 print(f" El producto {codigo}-{nombre} ha sido eliminado del carrito.")
                 break

  def mostrar(self):
      carrito = [producto for producto in self.productos]
      if carrito:
            for producto in carrito:
                print(producto)
      else:
            print("El carrito esta vacio.")

  def totalcom(self):
      totalcompra = sum(producto.precio * producto.stock for producto in self.productos)
      print(f"El costo total de la compra es: {totalcompra}")

  def descuentoprod(self,codigo,percent):
     for producto in self.productos:
          if producto.codigo == codigo:
              desprod = ((producto.precio * producto.stock)*(percent / 100))
              totdes = ((producto.precio*producto.stock) - desprod)
              print(f"El nuevo precio del producto con el descuento aplicado es de: {totdes}.")
              break
     else:
      print("El producto no se encuentra en el carrito.")

  def desctotal(self,percent):
      totalcompra = sum(producto.precio * producto.stock for producto in self.productos)
      descomp = (totalcompra * (percent / 100))
      finaltot = totalcompra - descomp
      print(f"Elcosto total de su compra con el descuento aplicado es: {finaltot}.")


carrito = carritodecompras()

while True:
  OPS= input("\n(1) Ver menu de opciones (0) Salir : ")
  if OPS=='1':
   print("\n OPCIONES:")
   print("\n(x) Ver productos disponibles en tienda\n(1) Añadir un producto al carrito\n(2) Eliminar un producto del carrito")
   print("(3) Mostrar los productos en el carrito\n(4) Calcular el total de la compra\n(5) Aplicar descuentos\n(0) Salir\n")
   print("NOTA: Si no ha añadido ningun producto al carrito las otras opciones no podrán ser ejecutadas.")
   P = input("\n¿Qué desea realizar?: ")

   if P =='1':
     while True:
      A= input("\n¿Desea añadir un producto al carrito? SI(a) NO(b): ")
      if A=='a':
        codigo = input("\nIngrese el codigo/identificador del producto: ")
        stock = int(input("Ingrese la cantidad que desea agregar al carrito: "))
        carrito.nuevo(codigo, stock)
      else:
       print("\nSe finalizó exitosamente el registro de su(s) producto(s) en el carrito de compras.")
       break
   elif P =='2':
     while True:
      E= input("\n¿Desea eliminar un producto al carrito? SI(a) NO(b): ")
      if E=='a':
        codigo = input("\nIngrese el codigo/identificador del producto que desea eliminar: ")
        carrito.eliminar(codigo)
      else:
        print("\nSe finalizó exitosamente la eliminación de su(s) producto(s) del carrito de compras.")
        break
   elif P =='3':
       carrito.mostrar()
   elif P == '4':
       carrito.totalcom()
   elif P == '5':
      pc= input("\n¿Como desde aplicar el descuento? por producto(a) o por total de compra(b): ")
      if pc=='a':
       codigo = input("\nIngrese el codigo/identificador del producto al que desea aplicar un descuento: ")
       percent= int(input("¿De qué porcentaje es el descuento?: "))
       carrito.descuentoprod(codigo,percent)
      else:
       percent= int(input("\n¿De qué porcentaje es el descuento?: "))
       carrito.desctotal(percent)
   elif P == '0':
        print("Se ha finalizado el programa")
        break
   elif P== 'x':
        print("Productos disponibles en la tienda:")
        for producto in productosdispo:
              print(producto)
   else:
    print("Entrada no valida.")
  else:
    print("Se ha finalizado el programa")
    break
```
